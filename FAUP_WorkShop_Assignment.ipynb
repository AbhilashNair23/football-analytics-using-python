**Problem Statement** : *Find Out the Best Replacement for Harry Kane at Tottenham?*


```python
# Import required libraries:
import json
import pandas as pd
from bs4 import BeautifulSoup
from urllib.request import urlopen
import io
import numpy as np
from copy import deepcopy
# Import plotting libraries
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
```

```python
# Scrape urls for 5 leagues

scrape_urls=["https://understat.com/league/EPL/2020",
             "https://understat.com/league/La_liga/2020",
             "https://understat.com/league/Bundesliga/2020",
             "https://understat.com/league/Serie_A/2020",
             "https://understat.com/league/Ligue_1/2020"]
```

```python
#Scraping top 10 players stats of each league

top_players_df = []
for url in scrape_urls:
  page_connect = urlopen(url)
  page_html = BeautifulSoup(page_connect, "html.parser")
  page_html.findAll(name="script")[3]
  json_raw_string = page_html.findAll(name="script")[3].text
  start_ind = json_raw_string.index("\\")
  stop_ind = json_raw_string.index("')")
  json_data = json_raw_string[start_ind:stop_ind]
  json_data = json_data.encode("utf8").decode("unicode_escape")
  players_json_df = pd.json_normalize(json.loads(json_data)).head(10)
  top_players_df.append(players_json_df)
top_players_df = pd.concat(top_players_df)
top_players_df
```

```python
# Reindex the dataframe
top_players_df.reset_index(drop=True, inplace=True)
```

```python
top_players_df["id"] = top_players_df["id"].astype(str).astype(int)
```

```python
# Create new fields based on per 90 calculation

per90Cols = ["goals", "xG", "assists", "xA", "shots", "key_passes", "npg", "npxG", "xGChain", "xGBuildup"]
for col in per90Cols:
 top_players_df[col + "Per90"] = top_players_df[col].astype(float).divide(top_players_df["time"].astype(float)).multiply(90)
```

```python
cols_for_radar = [i + "Per90" for i in per90Cols]
```

```python
# Find harry kane's data
top_players_df.loc[(top_players_df["id"] == 647), cols_for_radar].sum()
```

```python
# Harry Kane's Performance

fig = px.line_polar(top_players_df, 
                    r=top_players_df.loc[top_players_df["id"] == 647, cols_for_radar].sum(),
                    theta=cols_for_radar, line_close=True)
fig.update_traces(fill='toself')
fig.show()
```

```python
# Normalize the values
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
top_players_df[cols_for_radar] = scaler.fit_transform(top_players_df[cols_for_radar])
top_players_df[cols_for_radar].max()
```

```python
# Normalized Radar Plot for Kane's performance

fig = px.line_polar(top_players_df, r=top_players_df.loc[(top_players_df["id"] == 647), cols_for_radar].sum(),
                    theta=cols_for_radar, line_close=True)
fig.update_traces(fill='toself')
fig.show()
```

```python
# Normalized Radar Plot for all player with line polar

for i, row in top_players_df.iterrows():
    fig = px.line_polar(top_players_df, r=top_players_df.loc[(top_players_df["id"] == row["id"]), cols_for_radar].sum(),
                        theta=cols_for_radar, line_close=True,
                        title=row["player_name"])
    fig.update_traces(fill='toself')
    fig.show()
```

```python
max = top_players_df[cols_for_radar].max().max()
```

```python
players_id_names = top_players_df.drop_duplicates(subset=["id"])[["id", "player_name"]]
```

```python
# Finding the replacement for Harry Kane by comparing with every top 10 players of 5 leagues

for i, row in players_id_names.iterrows():
    if row["id"] == 647:
        continue
    print(row["player_name"])
    # Initiate the plotly go figure
    fig = go.Figure()
    # Add Radar plots for different players:
    fig.add_trace(go.Scatterpolar(
        r=top_players_df.loc[(top_players_df["id"] == 647), cols_for_radar].sum(),
        theta=cols_for_radar,
        fill='toself',
        name='Harry Kane'))
    fig.add_trace(go.Scatterpolar(
        r=top_players_df.loc[(top_players_df["id"] == row["id"]), cols_for_radar].sum(),
        theta=cols_for_radar,
        fill='toself',
        name=row["player_name"]))
    # Additional properties for the plot:
    fig.update_layout(
        title="Harry Kane vs " + row["player_name"],
    polar=dict(
        radialaxis=dict(
        visible=True,
        range=[0, max]
        )),
    showlegend=True
    )
    fig.show()
```
